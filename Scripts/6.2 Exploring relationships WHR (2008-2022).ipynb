## 3. Exploring Relationships

### Correlations:

#### Create a correlation heatmap

```python
# Create a subset excluding the "Country", "Region" and "Year" columns

sub = df[['Happiness Score', 'GDP per Capita','Social Support', 'Life Expectancy', 'Freedom', 'Charity', 'Corruption Perception', 'Positive Emotions', 'Negative Emotions']]
```

```python
sub
```

```python
# Create a subplot with matplotlib

f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.

corr = sns.heatmap(sub.corr(), annot = True, ax = ax) # The 'annot' argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

#### Legend:
    Purple equates to 0, or no relationship between the variables
    Dark purple and black equate to negative correlations
    Orange and light orange equate to positive correlations

#### Observation:

The important variable here is Happiness Score variable as it measures the country's overall happiness and well-being.

There are strong and positive correlations between Happiness Score and:

- GDP per Capita (0.78)
- Social Support (0.72)
- Life Expectancy (0.71)

We can also see a strong and positive relationship:

- Between GDP per Capita and Life Expectancy (0.82)

This correlation may highlights the importance of economic development in promoting improved health outcomes, and suggests that investments in economic growth and development may have positive spillover effects on population health.

There are medium-to-strong positive correlations between Happiness Score and:

- Fredom to make life choices (0.53)
- Positive Emotions (0.52)

On the other hand we can also see a moderate negative relationship between Happiness Score and:

- Corruption Perception (-0.43)
- Negative Emotions (-0.34)

This may suggests that as Corruption Perception and Negative Emotions (i.e. feelings of sadness, anger, worry, and stress) increase within a country, the Happiness Score tends to decrease, indicating lower levels of happiness and well-being among its citizens and vice-versa.

### Categorical Plots:

##### Creating a categorical variable that splits the "Happiness Score" column into categories

```python
# Use a histogram to visualize the distribution of the variables. 
# This way, we can determine sensible categories for the ranges. 
# We don't want to end up with too few observations in any one of the categories.
# The argument "kde" add a line that encompasses the distribution

sns.histplot(df['Happiness Score'], bins = 20, kde = True)
```

```python
df.loc[df['Happiness Score'] < 4, 'Happiness Score Category'] = 'Low score'
```

```python
df.loc[(df['Happiness Score'] >= 4) & (df['Happiness Score'] < 7), 'Happiness Score Category'] = 'Middle score'
```

```python
df.loc[df['Happiness Score'] >= 7, 'Happiness Score Category'] = 'High score'
```
